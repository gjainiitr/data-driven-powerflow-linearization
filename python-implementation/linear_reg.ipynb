```python
P = pd.read_csv('data/30P.csv', header = None)
Q = pd.read_csv('data/30Q.csv', header = None)
V = pd.read_csv('data/30V.csv', header = None)
Va = pd.read_csv('data/30Va.csv', header = None)
```

```python
print(P.shape)
print(Q.shape)
print(V.shape)
print(Va.shape)
```

```python
num_train = P.shape[0]
num_bus = P.shape[1]
```

```python
pi = 3.141592
X = P.join(Q, lsuffix='_P', rsuffix='_Q')
Y = Va.join(V, lsuffix='_V', rsuffix='_Va')
```

```python
print(X.shape, Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
# Ordinary Least Squares (OLS)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
ols = LinearRegression()
scores = cross_val_score(ols, X, Y, cv=5, scoring="neg_mean_absolute_error")
print("Mean error: ", scores.mean())
print("Standard deviation: ", scores.std())
```

```python
ols.fit(X, Y)
ols.coef_.shape
```

```python
Xva_Xv = ols.coef_
```

```python
import matplotlib.pyplot as plt
Xva_Xv.shape
plt.figure(figsize=(20,10))
plt.imshow(Xva_Xv, interpolation='none')
plt.show()
```

```python
# Forward linear regression
ols2 = LinearRegression()
scores = cross_val_score(ols2, Y, X, cv=5, scoring="neg_mean_absolute_error")
print("Mean error: ", scores.mean())
print("Standard deviation: ", scores.std())
```
